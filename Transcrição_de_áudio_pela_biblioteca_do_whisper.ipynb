```python
import whisper
# Carrega o modelo Whisper
model = whisper.load_model("base") # Escolha "base", "medium", "large", etc.

# Transcreve o áudio
result = model.transcribe("mapa territorio e imagens etnicas gravacao 1.mp3", task="transcribe")
```

```python
# Divide o texto em linhas de até 80 caracteres
formatted_text = "\n".join([result["text"][i:i+80] for i in range(0, len(result["text"]), 80)])
print(formatted_text)
```

```python
# Função que salva a transcrição no formato SRT
def save_as_srt(result, output_filename="transcription.srt"):
    segments = result["segments"]
    with open(output_filename, "w") as f:
        for i, segment in enumerate(segments):
            # Número da linha
            f.write(f"{i + 1}\n")
            # Intervalo de tempo no formato SRT
            start = segment["start"]
            end = segment["end"]
            f.write(f"{format_time(start)} --> {format_time(end)}\n")
            # Texto transcrito
            f.write(f"{segment['text']}\n\n")
# Função que formata o tempo em segundos para o formato SRT (hh:mm:ss,SSS)
def format_time(seconds):
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = int(seconds % 60)
    milliseconds = int((seconds % 1) * 1000)
    return f"{hours:02}:{minutes:02}:{seconds:02},{milliseconds:03}"

# Salva o arquivo no formato SRT
save_as_srt(result, "transcription.srt")
```
